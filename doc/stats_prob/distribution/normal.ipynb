```python
from scipy.stats import norm
```

Perfurm bottles are filled with the average volume of 150cc and the standard deviation of 2cc. What is the probability of a bottle containing more than 153cc?

```python
norm.sf(153, 150, 2)
```

Perfurm bottles are filled with the average voulme of 150cc and the standard deviation of 2cc. What percent of bottles will have a volume between 148cc and 152cc?

```python
norm.sf(148, 150, 2) - norm.sf(152, 150, 2)
```
